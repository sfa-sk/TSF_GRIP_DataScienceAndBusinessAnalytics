### **Libraries and Iris Dataset Imported**

```python
import sklearn.datasets as datasets
import pandas as pd
 
iris= datasets.load_iris()
x= pd.DataFrame(iris.data,columns=iris.feature_names)
x.head()
```

### **Iris dataset Target Printed**

```python
y= iris.target
print(y)
```

```python
x.info()
```

```python
x.describe()
```

### **Decision Tree Classifier Training**

```python
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier()
dec.fit(x,y)
```

## **Decision Tree Classifier Visualization**
Decision Tree Classifier is one of the most easy to understand, visualize and interpret Machine Learning Algorithm. 

### **Visualization Libraries installed**

```python
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
fig = plt.figure(figsize=(25,20))
vis = tree.plot_tree(dec, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
```
